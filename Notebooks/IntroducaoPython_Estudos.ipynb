```python
# if/else: controle de fluxo lógico
numero = input("Escreva um número de 1 à 10: ")
print(numero)
print(type(numero))

# o programa para no if se numero não tiver um número que possa
# ser convertido em inteiro!!!
if numero.isdigit(): # isdigit() é um método que toda string possui!

  # Indentação é o espaço que indica um bloco de código

  # Cast de tipo
  # o cast de tipo é converter um tipo em outro
  numero = int(numero)
  print(numero ** 2)

  # If dentro de outro If!!! :-O
  if numero == 5:
    print("O número é igual a 5")

    if numero == 6:
      print("O número é igual a 6")

# operador de comparação
# == avalia se temos duas variáveis iguais
# != avalia se temos duas variáveis diferentes

print(numero == 6)
```

```python
# Uso da função "def"
# graus Celsius 30ºC
# graus Farenheit 32ºF

def converte_temperatura(celsius):
  farenheit = celsius * 1.8 + 32 # converte celsius para farenheit
  return farenheit

f = converte_temperatura(30)
print("A temperatura", 30, "ºC em Farenheit é igual a", f , "Fº")
```
